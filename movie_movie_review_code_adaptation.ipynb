```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from lxml import html

chrome_options = Options()
chrome_options.add_argument("--headless") 

## These two just give headless browsing (= no browser actually pops up), which speeds things up. 
## Without them  my run times were often about 40s for each function call, whereas with them 
# it seems to drop to ~10 s per each function call

browser = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options = chrome_options)
```

```python
## TODO: Make a dictionary with the days too? May manually have to sort by earliest reviews and find the pages 
## which correspond to reviews after the awarding. So for instance, an entry make look as 'movie':"6" if page 6 
# is the last page with reviews before the awards

winners_2024 = ['anora', 'conclave', 'the-brutalist', 'emilia-perez', 'nickel-boys','wicked-2024','maria-2024']
## May want to take out Emilia Perez

winners_2023 = ['oppenheimer-2023', 'poor-things-2023', 'the-holdovers', 'american-fiction']

winners_2022 = ['everything-everywhere-all-at-once', 'all-quiet-on-the-western-front-2022', 'the-fabelmans', 'the-banshees-of-inisherin', 'women-talking', 'top-gun-maverick', 'elvis-2022']

winners_2021 = ['coda-2021', 'the-power-of-the-dog', 'west-side-story-2021', 'dont-look-up-2021', 'king-richard', 'tick-tick-boom-2021', 'dune-2021']

winners_2020 = ['mank', 'nomadland', 'the-trial-of-the-chicago-7', 'borat-subsequent-moviefilm', 'promising-young-woman', 'palm-springs-2020']

winners_2019 = ['parasite-2019', '1917', 'once-upon-a-time-in-hollywood', 'jojo-rabbit']

winners_2018 = ['green-book', 'roma-2018', 'black-panther', 'bohemian-rhapsody', 'eighth-grade', 'can-you-ever-forgive-me', 'the-favorite-2019', 'cold-war-2018']

winners_2017 = ['the-shape-of-water', 'three-billboards-outside-ebbing-missouri', 'lady-bird', 'get-out-2017', 'call-me-by-your-name', 'dunkirk-2017', 'i-tonya', 'blade-runner-2049']

winners_2016 = ['moonlight-2016', 'la-la-land', 'hidden-figures', 'arrival-2016', 'lion']

winners_2015 = ['spotlight', 'the-big-short', 'the-revenant-2015', 'the-martian', 'mad-max-fury-road']

winners_2014 = ['birdman-or-the-unexpected-virtue-of-ignorance', 'boyhood', 'the-grand-budapest-hotel', 'the-imitation-game']

award_winners = winners_2024 + winners_2023 + winners_2022 + winners_2021 + winners_2020 + winners_2019 + winners_2018 + winners_2017 + winners_2016 + winners_2015 + winners_2014
```

```python


def movie_ratings():

    # Open the virtual browser
    browser = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options = chrome_options)

    filename = 'movie_review_winner_reviews_all_times.csv'
    with open(filename, 'w+') as file_object:
        file_object.write("movie,.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5\n")


    with open(filename, 'w+') as file_object:

        for g in range(len(award_winners)):

            print(f"Opening " + award_winners[g] + "'s page...")
            pagename = 'https://letterboxd.com/csi/film/' + award_winners[g] + '/ratings-summary/'
            browser.get(pagename)
            print("Getting innerHTML...")
            innerHTML = browser.execute_script("return document.body.innerHTML")
            print("Creating tree...")
            tree = html.fromstring(innerHTML)
        
            # Grab the relevant data and put into lists
            print("parsing...")

            reviews_05 = tree.xpath('//li[@style="width: 15px; left: 0px"]/a/@title')
            reviews_1 = tree.xpath('//li[@style="width: 15px; left: 16px"]/a/@title')     
            reviews_15 = tree.xpath('//li[@style="width: 15px; left: 32px"]/a/@title') 
            reviews_2 = tree.xpath('//li[@style="width: 15px; left: 48px"]/a/@title')  
            reviews_25 = tree.xpath('//li[@style="width: 15px; left: 64px"]/a/@title') 
            reviews_3 = tree.xpath('//li[@style="width: 15px; left: 80px"]/a/@title')  
            reviews_35 = tree.xpath('//li[@style="width: 15px; left: 96px"]/a/@title')   
            reviews_4 = tree.xpath('//li[@style="width: 15px; left: 112px"]/a/@title')
            reviews_45 = tree.xpath('//li[@style="width: 15px; left: 128px"]/a/@title')  
            reviews_5 = tree.xpath('//li[@style="width: 15px; left: 144px"]/a/@title')

            print("Removing Commas...")

            print("Extracting Review Number...")            
            sr05 = reviews_05[0].replace(",","").split(" ")[0]          
            sr1 = reviews_1[0].replace(",","").split(" ")[0]
            sr15 = reviews_15[0].replace(",","").split(" ")[0]
            sr2 = reviews_2[0].replace(",","").split(" ")[0]
            sr25 = reviews_25[0].replace(",","").split(" ")[0]
            sr3 = reviews_3[0].replace(",","").split(" ")[0]
            sr35 = reviews_35[0].replace(",","").split(" ")[0]
            sr4 = reviews_4[0].replace(",","").split(" ")[0]
            sr45 = reviews_45[0].replace(",","").split(" ")[0]
            sr5 = reviews_5[0].replace(",","").split(" ")[0]

            print("Writing to File...")
            file_object.write(award_winners[g]+",")
            file_object.write(sr05+","+sr1+","+sr15+","+sr2+","+sr25+","+sr3+","+sr35+","+sr4+","+sr45+","+sr5+"\n")


        print(f"File {filename} written")
```

```python
movie_ratings()
```
